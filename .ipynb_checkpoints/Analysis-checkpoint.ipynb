```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock_data = pd.read_csv('results.csv')
```

# Data Preview

```python
print(stock_data)
```

# Hourly High Price Trend of Each Company

```python
for company in stock_data['company'].unique():
    plot_df = stock_data[stock_data['company'] == company]
    plt.plot(plot_df.hour, plot_df.hourly_high_price)
    plt.title(f"Hourly High Price Trend of '{company}'")
    plt.xlabel("Hour")
    plt.ylabel("Hourly High Price")
    plt.show()
```
